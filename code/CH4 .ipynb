```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/Zhou/Desktop/DoD修改2209/DoD_删后.csv',encoding='UTF-8')
```

```python
lst= df['Recipient Organization（消歧后）'].unique()
```

```python
res = {}
for i in range(len(lst)):
    df1 = df[df['Recipient Organization（消歧后）']==lst[i]]
    #print(df1)
    res[lst[i]]=df1['Anticipated Award Amount'].sum()
```

```python
res
```

```python
res_df = pd.DataFrame(pd.Series(res), columns=['sum'])
res_df
```

```python
res_df_sorted = res_df.sort_values(by = 'sum',ascending=False)[0:10]
res_df_sorted
```

```python
lst1 = []
for i in res_df_sorted.index:
    lst1.append(i)
lst1
```

```python
df2 = pd.read_csv('C:/Users/Zhou/Desktop/DoD修改2209/DoD_删后.csv')
```

```python
df3 = df2[df2['Recipient Organization（消歧后）'] == lst1[0]]
for i in range(len(lst1)):
    if(i==0):
        continue
    else:
        df_temp = df2[df2['Recipient Organization（消歧后）'] == lst1[i]]
        df3 = df3.append(df_temp)
        
len(df3)
```

```python
df3.to_csv('Recipient Organization（消歧后）.csv')
```

```python

from stylecloud import gen_stylecloud
 
# 读取文件
pd_data = pd.read_csv('Recipient Organization（消歧后）.csv')
exist_col = pd_data.dropna()  # 删除空行

pd_data
```

```python
 
# 读取内容
text = exist_col['Award Abstract'].tolist()
 
# 切割分词
wordlist = jieba.cut_for_search(''.join(text))
result = ' '.join(wordlist)

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
custom_stopwords=['result','one','used','using','work','study','studies','provide','well','develop','proposed','propose','approach','objective','methods','including','also','project','鈥','钬','based','use','problem','however','would','in','the','for','will','be','or','a','and','on','as','to','of','this','that','we','from','such','current','measurement','analysis','method','new', 'data', 'other', 'system', 'model', 'novel','systems','materials','research','models','program','design','field','structure']
stop_words.extend(custom_stopwords)
             
wc=gen_stylecloud(text=result,
                icon_name='fas fa-graduation-cap',#https://fontawesome.com/icons/graduation-cap?s=solid&f=classic  fas fa- 加上标题
                font_path='msyh.ttc',
                background_color='white',
                output_name='661.jpg',
                size=1000,
                custom_stopwords=stop_words
                )
print('绘图成功！')
```
